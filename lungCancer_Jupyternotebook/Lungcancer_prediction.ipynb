```python
import pandas as pd
dataset = pd.read_csv("lungcancer.csv")
X=dataset.iloc[:,:-1]
Y=dataset.iloc[:,-1]
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score,confusion_matrix,ConfusionMatrixDisplay

import numpy as np
X_train,X_test,Y_train,Y_test=train_test_split(X,Y,train_size=.90)
model=RandomForestClassifier(random_state=48)
model.fit(X_train,Y_train)
Y_predicted=model.predict(X_test)

rf_cm=confusion_matrix(Y_predicted,Y_test)
rf_cm


```

```python
 accuracy_score(Y_predicted,Y_test)
rf_accuracy=accuracy_score(Y_predicted,Y_test)
rf_accuracy
```

```python
cm_display=ConfusionMatrixDisplay(confusion_matrix=rf_cm,display_labels=["Cancer","Normal"])
cm_display.plot()
```

```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import accuracy_score, confusion_matrix, ConfusionMatrixDisplay


dataset = pd.read_csv("lungcancer.csv")


X = dataset.iloc[:, :-1]
Y = dataset.iloc[:, -1]


X_train, X_test, Y_train, Y_test = train_test_split(X, Y, train_size=0.90, random_state=48)


tree_classifier = DecisionTreeClassifier(random_state=48)  # You can customize hyperparameters here


tree_classifier.fit(X_train, Y_train)


Y_predicted = tree_classifier.predict(X_test)
```
